```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x, y)
plt.show()
```

```python
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])

plt.plot(x, y)
plt.show()
```

```python
y = np.array([3, 8, 1, 10])

plt.plot(y, marker = 'X')
plt.show()
```

```python
y = np.array([3, 8, 1, 10])

plt.plot(y, 'X:r')
plt.show()
```

```python
y = np.array([3, 8, 1, 10])

plt.plot(y, marker = 'X', ms = 20)
plt.show()
```

```python
x = np.array([45, 20, 20, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(x, labels = mylabels, startangle = 90)
plt.show()
```
